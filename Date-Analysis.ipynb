```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('csv/metadata_text_merged.csv')
# Remove rows with empty text column for now
df = df.dropna(subset=['text'])
# Add year column
df['Year'] = pd.to_datetime(df['Date'], format='%d/%m/%Y', errors='coerce').dt.year
# Add month column
df['Month'] = pd.to_datetime(df['Date'], format='%d/%m/%Y', errors='coerce').dt.month
```

```python
# Visualize counts of letters over time
plt.figure()
plt.title('Count by Year')
df.groupby('Year').count()['UID'].plot()
```
